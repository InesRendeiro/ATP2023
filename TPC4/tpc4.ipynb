```python
import random

def criaLista(N):
    lista=[]
    while N>0:
        n=random.randint(1,100)
        lista.append(n)
        N=N-1
    return lista 

def leLista(N):
    lista=[]
    i=0
    while i<N:
        n=int(input("introduza um numero"))
        lista.append(n)
        i=i+1   
    return lista

def somaLista(lista):
    soma=0
    for a in lista:
        soma=soma+a
    return soma

def quadrados(lista):
    lista2=[]
    for a in lista:
        lista2.append(a*a)
    return lista2

def mediaLista(lista):
    comp=len(lista)
    media=somaLista(lista)/comp
    return media

def maiorLista(lista):
    maior=lista[0]
    for a in lista:
        l=0
        while l<len(lista):
            if maior<lista[l]:
                maior=a
            l=l+1
    return maior

def menorLista(lista):
    menor=lista[0]
    for a in lista:
        l=0
        while l<len(lista):
            if menor>lista[l]:
                menor=a
            l=l+1
    return menor


def estaOrdenadaC(lista):
    valor=True
    l=1
    while l<len(lista):
        if lista[l] < lista[l-1]:
            valor=False
        l=l+1
    return valor

def estaOrdenadaD(lista):
    valor=True
    l=1
    while l<len(lista):
        if lista[l] > lista[l-1]:
            valor=False
        l=l+1
    return valor

def indiceDe(lista, elem):
    valor=-1
    l=0
    while l<len(lista):
        if lista[l]==elem:
            valor=l
        l=l+1
    return valor

def menu():
    print('''qual vai ser o operação? 
(0)sair, 
(1)criar Lista, 
(2)le lista, 
(3)soma, 
(4)média, 
(5)maior, 
(6)menor,
(7)Ordem crescente, 
(8)Ordem decrescente,
(9)procurar elem''')

menu()
func=int(input("qual vai ser a função que quer?"))
res=[]

while func !=0:
    if func==1:
        res.clear()
        res=criaLista(int(input("Qual o tamanho da lista?")))
        print("")
        print(res)
    if func==2:
        res.clear()
        res=leLista(int(input("Qual o tamanho da lista?")))
        print("")
        print(res)
    if func==3:
        print("")
        print("A soma  da sua lista é:",somaLista(res))
    if func==4:
        print("")
        print("A média  da sua lista é:",mediaLista(res))
    if func==5:
        print("")
        print("O maior elemento da sua lista é:",maiorLista(res))
    if func==6:
        print("")
        print("O menor elemento da sua lista é:",menorLista(res))
    if func==7:
        print("")
        print("A lista está ordenada na forma crescente:",estaOrdenadaC(res))
    if func==8:
        print("")
        print("A lista está ordenadade na forma decrescente:",estaOrdenadaD(res))
    if func==9:
        pos= int(indiceDe(res,int(input("Qual o elemento que procura?"))))
        if pos==-1:
            print("")
            print("O elemento não se encontra na lista.")
        else:
            print("")
            print("O elemento encontra-se na(s)",pos,"posição(ões).")
    func=int(input("O que pretende fazer?"))
print("")
print("Fechaste o programa")

```
