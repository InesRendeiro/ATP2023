```python
#Cinema = [Sala]
#Sala = [nlugares, Vendidos, filme]
#nlugares = Int
#filme = String 
#Vendidos = [Int]
def menu():
    print("""Bem-vindo aos cinemas Nos 
Menu:
    (1) Consultar todos os filmes em exibição no cinema
    (2) Verificar se um lugar está disponível
    (3) Vender bilhete
    (4) Consultar quantos lugares ainda estão disponíveis na sala
    (5) Criar uma nova sala
    (0) Sair """)
    return


def listar(cinema):
    for x in cinema:
        print("O cinema tem em exibição:", x[2])  


def inserirSala(cinema, sala):
    if cinema==[]:
        cinema.append(sala)
    else:
        for x in cinema:
            if sala[2]!=x[2]:
                cinema.append(sala)
    return cinema

def disponivel(cinema,filme,lugar):
    valor=False
    for x in cinema:
        if filme==x[2]:
            if lugar not in x[1]:
                valor=True
    return valor

def listardisponibilidades(cinema):
    for x in cinema: 
        if len(x[1])==x[0]:
            print(x[2],"está ocupado!")
        else:
            print(x[2],"está livre e tem",x[0]-len(x[1]),"lugares!") 
    return cinema

def vendebilhete( cinema, filme, lugar):
    for x in cinema:
        if filme==x[2]:
            if lugar not in x[1]:
                x[1].append(lugar)
            x[0]=x[0]-len(x[1])
    return cinema

def criaSala (cinema,filme,lugar):
    cinema.append([lugar,[],filme])
    return cinema

sala1 = [150, [], "Twilight"]
sala2 = [200, [], "Hannibal"]
cinema1 = []

cinema1 = inserirSala(cinema1,sala1)
cinema1 = inserirSala(cinema1,sala2)


menu()
function=int(input("O que pretende fazer?"))

while function !=0:
    if function == 1:
        print("")
        listar(cinema1)

    if function == 2:
        print("")
        if disponivel (cinema1,input("Insira o nome do filme:"),int(input("Insere o número do lugar:"))) == True:
            print ("O bilhete está disponível!")
        else:
            print ("O bilhete não está disponível!")

    if function == 3:
        print("")
        print(vendebilhete (cinema1,input("Insira o nome do filme"), int(input("Insira o número do lugar:"))))

    if function == 4:
        print("")
        print(listardisponibilidades (cinema1))

    if function == 5:
        print("")
        cinema1.append(criaSala(cinema1,input("Nome do novo filme"),int(input("Nº de lugares da nova sala"))))
        print(cinema1)

        
    function = int(input("O que pretende consultar de seguida?"))
print("")
print("Obrigado pela sua preferência!")
```
